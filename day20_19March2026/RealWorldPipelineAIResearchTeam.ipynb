Write a comprehensive market report on EV industry trends in India for 2025.

Step-by-Step Execution Plan
1️⃣ Define Scope & Objectives

What to cover:

Market size & growth

Key players (Tata, Ola Electric, etc.)

Government policies (FAME, subsidies)

Consumer trends

Infrastructure (charging stations)

Output format: structured report

2️⃣ Data Collection (Search Phase)

Gather latest data from:

Industry reports

News articles

Government sources

Focus on:

2023–2025 projections

EV adoption rates in India

3️⃣ Data Analysis (Analyst Phase)

Interpret collected data:

Growth trends (EV sales, CAGR)

Segment breakdown:

2-wheelers

3-wheelers

4-wheelers

Key drivers:

Fuel prices

Policy incentives

Identify patterns + insights

4️⃣ Report Drafting (Writer Phase)

Structure the report:

Introduction

Market Overview

Key Trends

Challenges

Opportunities

Future Outlook (2025)

Convert analysis into readable content

5️⃣ Quality Check (QA Phase)

Verify:

Facts & consistency

Errors

Missing insights

6️⃣ Final Output

Deliver:

Clean, structured market report

Clear 2025 outlook

####Simple Flow Summary
Goal → Collect Data → Analyze → Write → Validate → Final Report

```python
import requests
from google.colab import userdata

API_KEY = userdata.get('groq_api_key')

# Debug check
if not API_KEY:
    raise ValueError("❌ API key not found in Colab secrets")

print(" API key loaded:", API_KEY[:10], "...")  # shows first few chars


# ---------------- LLM CALL ---------------- #
def call_llm(prompt):
    url = "https://api.groq.com/openai/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "llama-3.1-8b-instant",
        "messages": [
            {"role": "system", "content": "You are a precise business analyst."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.3,
        "max_tokens": 800
    }

    try:
        res = requests.post(url, headers=headers, json=data, timeout=30)

        if res.status_code != 200:
            print("Status:", res.status_code)
            print("Response:", res.text)

        res.raise_for_status()
        return res.json()["choices"][0]["message"]["content"]

    except Exception as e:
        return f"Error: {e}"


# ---------------- AGENTS ---------------- #

class SearchAgent:
    def run(self, goal):
        print("\n[Search Agent] collecting data...")

        prompt = f"""
Goal: {goal}

Find key facts, stats, and recent trends about EV industry in India (2025 focus).
Keep it concise.
"""
        result = call_llm(prompt)
        print(result)
        return result


class AnalystAgent:
    def run(self, data):
        print("\n[Analyst Agent] analyzing...")

        prompt = f"""
Analyze this EV market data:

{data}

Extract trends, insights, growth drivers, and challenges.
"""
        result = call_llm(prompt)
        print(result)
        return result


class WriterAgent:
    def run(self, analysis):
        print("\n[Writer Agent] writing report...")

        prompt = f"""
Write a structured market report based on:

{analysis}

Include:
- Introduction
- Market trends
- Challenges
- Opportunities
- 2025 outlook
"""
        result = call_llm(prompt)
        print(result)
        return result


class QAAgent:
    def run(self, report):
        print("\n[QA Agent] reviewing report...")

        prompt = f"""
Review this report for clarity, correctness, and flow:

{report}

Fix issues and improve quality.
"""
        result = call_llm(prompt)
        print(result)
        return result


# ---------------- ORCHESTRATOR ---------------- #

class Orchestrator:
    def run(self, goal):
        print(f"\n[Orchestrator] Goal: {goal}")

        search = SearchAgent()
        analyst = AnalystAgent()
        writer = WriterAgent()
        qa = QAAgent()

        data = search.run(goal)
        analysis = analyst.run(data)
        draft = writer.run(analysis)
        final = qa.run(draft)

        return final


# ---------------- RUN ---------------- #

if __name__ == "__main__":
    goal = "Write a comprehensive market report on EV industry trends in India for 2025"
    orchestrator = Orchestrator()
    final_report = orchestrator.run(goal)
    print("\n\n✅ FINAL REPORT:\n",final_report)
```
