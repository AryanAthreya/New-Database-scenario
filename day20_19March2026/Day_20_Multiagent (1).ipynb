```python
# 🎯 Example 1: Sequential Multi-Agent System (Pipeline)
# 🧠 Scenario

# “A travel company has different employees (agents):

# Planner → decides steps

# Flight Agent → finds flights

# Weather Agent → checks weather

# Decision Agent → gives final answer”

# ================================
# AGENT 1: PLANNER
# ================================
def planner_agent(user_query):
    print("\n[Planner Agent] Creating plan...")
    return ["flight", "weather", "decision"]


# ================================
# AGENT 2: FLIGHT AGENT
# ================================
def flight_agent():
    print("\n[Flight Agent] Fetching flights...")
    return [
        {"airline": "IndiGo", "price": 4500},
        {"airline": "Air India", "price": 5200}
    ]


# ================================
# AGENT 3: WEATHER AGENT
# ================================
def weather_agent():
    print("\n[Weather Agent] Checking weather...")
    return {"condition": "Clear", "temp": 28}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(flights, weather):
    print("\n[Decision Agent] Making decision...")

    cheapest = min(flights, key=lambda x: x["price"])

    if weather["condition"] == "Rain":
        return "Avoid travel due to bad weather"

    return f"Book {cheapest['airline']} at ₹{cheapest['price']}"


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def travel_multi_agent(user_query):
    print("User Query:", user_query)

    plan = planner_agent(user_query)

    flights = None
    weather = None

    for step in plan:
        if step == "flight":
            flights = flight_agent()

        elif step == "weather":
            weather = weather_agent()

        elif step == "decision":
            result = decision_agent(flights, weather)

    return result


# RUN
response = travel_multi_agent("Plan my trip Delhi to Mumbai")
print("\nFinal Answer:", response)
```

```python
# Example 2: Manager–Worker Multi-Agent System
# 🧠 Scenario

# “Now instead of fixed flow, we introduce a Manager Agent
# that assigns tasks dynamically to worker

# ================================
# WORKER AGENTS
# ================================
def flight_agent():
    print("[Flight Agent] Working...")
    return [{"airline": "IndiGo", "price": 4500},
            {"airline": "Air India", "price": 5200}]


def weather_agent():
    print("[Weather Agent] Working...")
    return {"condition": "Clear", "temp": 28}


# ================================
# MANAGER AGENT
# ================================
def manager_agent(user_query):
    print("\n[Manager Agent] Analyzing task...")

    tasks = []

    if "trip" in user_query.lower():
        tasks = ["flight", "weather"]

    return tasks


# ================================
# EXECUTION
# ================================
def run_system(user_query):
    print("User Query:", user_query)

    tasks = manager_agent(user_query)

    results = {}

    for task in tasks:
        if task == "flight":
            results["flights"] = flight_agent()

        elif task == "weather":
            results["weather"] = weather_agent()

    # Final decision
    cheapest = min(results["flights"], key=lambda x: x["price"])

    return f"Manager Decision: Book {cheapest['airline']} at ₹{cheapest['price']}"


# RUN
response = run_system("Plan my trip")
print("\nFinal Answer:", response)
```

practice 1 - Scenario
“A hospital uses different employees (agents) to handle patient care in sequence.”

================================
AGENT 1: Intake Agent (Planner)
- Collects patient symptoms and history
- Decides which steps are needed (tests, consultations, etc.)

================================
AGENT 2: Diagnostic Agent
- Orders lab tests or scans
- Interprets results and identifies possible conditions

================================
AGENT 3: Treatment Agent
- Suggests treatment options (medication, therapy, surgery)
- Considers patient preferences and medical guidelines

================================
AGENT 4: Decision Agent
- Reviews all inputs (history, diagnostics, treatment options)
- Provides the final recommendation to the patient

```python
import time

# ================================
# AGENT 1: INTAKE AGENT (PLANNER)
# ================================
def intake_agent(patient_symptoms, patient_history):
    print("\n[Intake Agent] Collecting patient info and planning next steps...")
    # Simulate collecting data and deciding a plan
    plan = []
    if "fever" in patient_symptoms.lower() or "cough" in patient_symptoms.lower():
        plan.append("diagnostics")
    if "pain" in patient_symptoms.lower() and "injury" in patient_history.lower():
        plan.append("diagnostics")
    plan.append("treatment")
    plan.append("decision")

    patient_info = {
        "symptoms": patient_symptoms,
        "history": patient_history,
        "plan": plan
    }
    time.sleep(1)
    return patient_info


# ================================
# AGENT 2: DIAGNOSTIC AGENT
# ================================
def diagnostic_agent(patient_info):
    print("[Diagnostic Agent] Ordering tests and interpreting results...")
    diagnostics = {}
    if "diagnostics" in patient_info["plan"]:
        # Simulate ordering tests
        if "fever" in patient_info["symptoms"].lower():
            diagnostics["lab_test"] = "positive for virus"
        elif "pain" in patient_info["symptoms"].lower():
            diagnostics["xray_result"] = "minor fracture"
        else:
            diagnostics["lab_test"] = "negative"

    time.sleep(1)
    return diagnostics


# ================================
# AGENT 3: TREATMENT AGENT
# ================================
def treatment_agent(patient_info, diagnostics_results):
    print("[Treatment Agent] Suggesting treatment options...")
    treatment_options = []
    if "treatment" in patient_info["plan"]:
        if "positive for virus" in diagnostics_results.get("lab_test", "").lower():
            treatment_options.append("Antiviral medication")
            treatment_options.append("Rest and fluids")
        elif "minor fracture" in diagnostics_results.get("xray_result", "").lower():
            treatment_options.append("Splinting")
            treatment_options.append("Pain medication")
        else:
            treatment_options.append("Observation and follow-up")

    time.sleep(1)
    return treatment_options


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(patient_info, diagnostics_results, treatment_options):
    print("[Decision Agent] Reviewing all inputs and providing recommendation...")
    final_recommendation = {
        "patient_summary": {
            "symptoms": patient_info["symptoms"],
            "history": patient_info["history"]
        },
        "diagnostics": diagnostics_results,
        "treatment_options": treatment_options,
        "recommendation": ""
    }

    if "positive for virus" in diagnostics_results.get("lab_test", "").lower():
        final_recommendation["recommendation"] = "Prescribe antiviral medication and recommend rest. Schedule a follow-up in 3 days."
    elif "minor fracture" in diagnostics_results.get("xray_result", "").lower():
        final_recommendation["recommendation"] = "Apply a splint, prescribe pain medication, and refer to orthopedics."
    elif not treatment_options:
         final_recommendation["recommendation"] = "No immediate treatment required, advise on symptom management and follow up if symptoms worsen."
    else:
        final_recommendation["recommendation"] = "Discuss treatment options with the patient and provide care instructions."

    time.sleep(1)
    return final_recommendation


# ================================
# MAIN HOSPITAL MULTI-AGENT SYSTEM
# ================================
def hospital_multi_agent(patient_query):
    print(f"User Query: {patient_query}")
    # Assume patient_query contains symptoms and history
    # For simplicity, we'll extract directly here
    symptoms_start = patient_query.find("symptoms:")
    history_start = patient_query.find("history:")

    patient_symptoms = patient_query[symptoms_start + len("symptoms:"):history_start].strip()
    patient_history = patient_query[history_start + len("history:"):].strip()

    # Step 1: Intake Agent
    patient_info = intake_agent(patient_symptoms, patient_history)

    diagnostics_results = {}
    treatment_options = []

    for step in patient_info["plan"]:
        if step == "diagnostics":
            # Step 2: Diagnostic Agent
            diagnostics_results = diagnostic_agent(patient_info)
        elif step == "treatment":
            # Step 3: Treatment Agent
            treatment_options = treatment_agent(patient_info, diagnostics_results)
        elif step == "decision":
            # Step 4: Decision Agent
            final_recommendation = decision_agent(patient_info, diagnostics_results, treatment_options)

    return final_recommendation


# RUN SCENARIOS
print("\n--- Scenario 1: Patient with fever and cough ---")
response1 = hospital_multi_agent("Patient symptoms: high fever, cough. Patient history: no significant past medical history.")
print("\nFinal Hospital Recommendation:")
for key, value in response1.items():
    print(f"  {key}: {value}")

print("\n--- Scenario 2: Patient with severe wrist pain after a fall ---")
response2 = hospital_multi_agent("Patient symptoms: severe wrist pain. Patient history: fell off bicycle an hour ago.")
print("\nFinal Hospital Recommendation:")
for key, value in response2.items():
    print(f"  {key}: {value}")

print("\n--- Scenario 3: Routine check-up ---")
response3 = hospital_multi_agent("Patient symptoms: none. Patient history: annual check-up.")
print("\nFinal Hospital Recommendation:")
for key, value in response3.items():
    print(f"  {key}: {value}")
```

```python
# Practice 1 - Scenario: Hospital Multi-Agent System (Gemini 1.5 Flash Powered)
# This version replaces the deterministic logic of the agents with calls to the Gemini 1.5 Flash API.

import google.generativeai as genai
import os
import time
import json

# Configure your Gemini API key
# You can get an API key from Google AI Studio: https://aistudio.google.com/app/apikey
# It's recommended to store your API key securely, e.g., in environment variables or Colab Secrets.
# For this example, you can directly paste it or use os.getenv.
# For Colab Secrets, use: from google.colab import userdata; API_KEY = userdata.get('GEMINI_API_KEY')

# API_KEY = "YOUR_GEMINI_API_KEY" # Replace with your actual API key
# genai.configure(api_key=API_KEY)

# --- Placeholder for API Key ---
# Since I cannot access your environment or prompt for a key,
# please uncomment and set your API key as shown above.
# For demonstration purposes, I will use a dummy key and handle potential errors.

try:
    # Attempt to load from Colab secrets or environment variable
    from google.colab import userdata
    API_KEY = userdata.get('google_api_key')
except ImportError:
    API_KEY = os.getenv('GEMINI_API_KEY')

if not API_KEY:
    print("Warning: GEMINI_API_KEY not found. Using a dummy key. Please configure your API key.")
    API_KEY = "dummy_key"

genai.configure(api_key=API_KEY)

# Initialize the Gemini model
# Using 'gemini-1.5-flash' for faster responses
model = genai.GenerativeModel('gemini-2.5-flash')

def call_gemini(prompt, temperature=0.2):
    """Helper function to call the Gemini API"""
    try:
        response = model.generate_content(
            prompt,
            generation_config=genai.GenerationConfig(
                temperature=temperature,
                response_mime_type="application/json"
            )
        )
        return json.loads(response.text)
    except Exception as e:
        print(f"Error calling Gemini API: {e}")
        print(f"Prompt: {prompt}")
        return {"error": str(e), "message": "Could not get response from Gemini."}

# ================================
# AGENT 1: INTAKE AGENT (PLANNER)
# ================================
def intake_agent_gemini(patient_symptoms, patient_history):
    print("\n[Intake Agent (Gemini)] Collecting patient info and planning next steps...")
    prompt = f"""You are an Intake Agent in a hospital. Your task is to collect patient symptoms and history, and then propose a plan for their care. The plan should include 'diagnostics', 'treatment', and 'decision' steps as needed.

Patient Symptoms: {patient_symptoms}
Patient History: {patient_history}

Based on this information, provide a JSON object with the following keys:
'symptoms': string (the provided symptoms)
'history': string (the provided history)
'plan': list of strings (e.g., ["diagnostics", "treatment", "decision"])
"""
    response_data = call_gemini(prompt)
    if "error" in response_data:
        # Fallback to a default plan if API call fails
        plan = ["diagnostics", "treatment", "decision"] # Default if Gemini fails
        print("Using fallback plan due to Gemini error.")
        return {"symptoms": patient_symptoms, "history": patient_history, "plan": plan}

    print("Gemini Intake Agent Response:", response_data)
    time.sleep(1)
    return response_data


# ================================
# AGENT 2: DIAGNOSTIC AGENT
# ================================
def diagnostic_agent_gemini(patient_info):
    print("[Diagnostic Agent (Gemini)] Ordering tests and interpreting results...")
    if "diagnostics" not in patient_info["plan"]:
        return {}

    prompt = f"""You are a Diagnostic Agent. Based on the patient's symptoms and history, and the overall care plan, suggest potential diagnostic tests and their likely results. If no specific tests are needed, indicate that.

Patient Symptoms: {patient_info['symptoms']}
Patient History: {patient_info['history']}
Care Plan: {patient_info['plan']}

Provide a JSON object with potential diagnostic results. For example:
{{ "lab_test": "positive for virus" }} or {{ "xray_result": "minor fracture" }} or {{}}
"""
    response_data = call_gemini(prompt)
    if "error" in response_data:
        print("Using fallback diagnostics due to Gemini error.")
        # Fallback for diagnostics
        if "fever" in patient_info["symptoms"].lower():
            return {"lab_test": "positive for virus (simulated)"}
        elif "pain" in patient_info["symptoms"].lower():
            return {"xray_result": "minor fracture (simulated)"}
        else:
            return {"lab_test": "negative (simulated)"}

    print("Gemini Diagnostic Agent Response:", response_data)
    time.sleep(1)
    return response_data


# ================================
# AGENT 3: TREATMENT AGENT
# ================================
def treatment_agent_gemini(patient_info, diagnostics_results):
    print("[Treatment Agent (Gemini)] Suggesting treatment options...")
    if "treatment" not in patient_info["plan"]:
        return []

    prompt = f"""You are a Treatment Agent. Based on the patient's symptoms, history, and diagnostic results, suggest appropriate treatment options.

Patient Symptoms: {patient_info['symptoms']}
Patient History: {patient_info['history']}
Diagnostic Results: {json.dumps(diagnostics_results)}

Provide a JSON object where the key is 'treatment_options' and the value is a list of strings, e.g., {{ "treatment_options": ["Antiviral medication", "Rest and fluids"] }}
"""
    response_data = call_gemini(prompt)
    if "error" in response_data:
        print("Using fallback treatment options due to Gemini error.")
        # Fallback for treatment
        if "positive for virus" in str(diagnostics_results).lower():
            return ["Antiviral medication (simulated)", "Rest and fluids (simulated)"]
        elif "minor fracture" in str(diagnostics_results).lower():
            return ["Splinting (simulated)", "Pain medication (simulated)"]
        else:
            return ["Observation and follow-up (simulated)"]

    print("Gemini Treatment Agent Response:", response_data)
    time.sleep(1)
    return response_data.get("treatment_options", [])


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent_gemini(patient_info, diagnostics_results, treatment_options):
    print("[Decision Agent (Gemini)] Reviewing all inputs and providing recommendation...")
    if "decision" not in patient_info["plan"]:
        return {"recommendation": "No final decision step in plan.", "summary": patient_info, "diagnostics": diagnostics_results, "treatment": treatment_options}

    prompt = f"""You are a Decision Agent. Review all available information (patient summary, diagnostics, treatment options) and provide a concise final recommendation to the patient.

Patient Summary: {json.dumps(patient_info)}
Diagnostic Results: {json.dumps(diagnostics_results)}
Treatment Options: {json.dumps(treatment_options)}

Provide a JSON object with the following keys:
'recommendation': string (the final recommendation)
'patient_summary': object (the patient_info passed)
'diagnostics': object (the diagnostics_results passed)
'treatment_options': list of strings (the treatment_options passed)
"""
    response_data = call_gemini(prompt)
    if "error" in response_data:
        print("Using fallback recommendation due to Gemini error.")
        # Fallback for decision
        return {
            "patient_summary": patient_info,
            "diagnostics": diagnostics_results,
            "treatment_options": treatment_options,
            "recommendation": "Could not get a recommendation from Gemini. Advise on symptom management and follow up if symptoms worsen (simulated)."
        }

    print("Gemini Decision Agent Response:", response_data)
    time.sleep(1)
    return response_data


# ================================
# MAIN HOSPITAL MULTI-AGENT SYSTEM (Gemini-Powered)
# ================================
def hospital_multi_agent_gemini(patient_query):
    print(f"\nUser Query: {patient_query}")

    symptoms_start = patient_query.find("symptoms:")
    history_start = patient_query.find("history:")

    patient_symptoms = patient_query[symptoms_start + len("symptoms:"):history_start].strip()
    patient_history = patient_query[history_start + len("history:"):].strip()

    # Step 1: Intake Agent (Gemini)
    patient_info = intake_agent_gemini(patient_symptoms, patient_history)

    diagnostics_results = {}
    treatment_options = []
    final_recommendation = {}

    # Ensure plan is a list, even if Gemini returned an error or different format
    plan = patient_info.get("plan", []) if isinstance(patient_info.get("plan"), list) else ["diagnostics", "treatment", "decision"]

    for step in plan:
        if step == "diagnostics":
            # Step 2: Diagnostic Agent (Gemini)
            diagnostics_results = diagnostic_agent_gemini(patient_info)
        elif step == "treatment":
            # Step 3: Treatment Agent (Gemini)
            treatment_options_response = treatment_agent_gemini(patient_info, diagnostics_results)
            # Ensure treatment_options is a list
            if isinstance(treatment_options_response, list):
                treatment_options = treatment_options_response
            else: # If Gemini returns a dict (e.g., error) or other, try to extract or default
                treatment_options = treatment_options_response.get("treatment_options", []) if isinstance(treatment_options_response, dict) else []

        elif step == "decision":
            # Step 4: Decision Agent (Gemini)
            final_recommendation = decision_agent_gemini(patient_info, diagnostics_results, treatment_options)

    return final_recommendation


# RUN SCENARIOS (Gemini-Powered)
print("\n--- Scenario 1 (Gemini): Patient with fever and cough ---")
response1_gemini = hospital_multi_agent_gemini("Patient symptoms: high fever, cough. Patient history: no significant past medical history.")
print("\nFinal Hospital Recommendation (Gemini):")
for key, value in response1_gemini.items():
    print(f"  {key}: {value}")

print("\n--- Scenario 2 (Gemini): Patient with severe wrist pain after a fall ---")
response2_gemini = hospital_multi_agent_gemini("Patient symptoms: severe wrist pain. Patient history: fell off bicycle an hour ago.")
print("\nFinal Hospital Recommendation (Gemini):")
for key, value in response2_gemini.items():
    print(f"  {key}: {value}")

print("\n--- Scenario 3 (Gemini): Routine check-up ---")
response3_gemini = hospital_multi_agent_gemini("Patient symptoms: none. Patient history: annual check-up.")
print("\nFinal Hospital Recommendation (Gemini):")
for key, value in response3_gemini.items():
    print(f"  {key}: {value}")
```

```python
# Practice 1 - Scenario: Hospital Multi-Agent System (QROQ Powered - Simulated)
# This version illustrates how to adapt the agents for a hypothetical QROQ API.

import os
import time
import json

# Configure your QROQ API key
# It's recommended to store your API key securely, e.g., in environment variables or Colab Secrets.
# For Colab Secrets, use: from google.colab import userdata; QROQ_API_KEY = userdata.get('QROQ_API_KEY')

# QROQ_API_KEY = "YOUR_QROQ_API_KEY" # Replace with your actual API key

try:
    from google.colab import userdata
    QROQ_API_KEY = userdata.get('groq_api_key')
except ImportError:
    QROQ_API_KEY = os.getenv('QROQ_API_KEY')

if not QROQ_API_KEY or QROQ_API_KEY == "YOUR_QROQ_API_KEY":
    print("Warning: QROQ_API_KEY not found or not set. Using a dummy key for simulation.")
    QROQ_API_KEY = "dummy_qroq_key"

# --- Simulated QROQ API Call ---
# In a real scenario, you would import a QROQ client library and make API calls here.
# For demonstration, we'll simulate a JSON response based on the prompt's intent.

def call_qroq(prompt, temperature=0.2):
    """Helper function to simulate calling a QROQ API and returning JSON"""
    print(f"[Simulated QROQ API Call] Processing prompt... (Temp: {temperature})")
    # In a real implementation, you'd use a QROQ client library here:
    # client = qroq_library.QroqClient(api_key=QROQ_API_KEY)
    # response = client.chat.completions.create(
    #    model="your-qroq-model",
    #    response_format={ "type": "json_object" },
    #    messages=[
    #        {"role": "user", "content": prompt}
    #    ]
    # )
    # return json.loads(response.choices[0].message.content)

    # For this simulation, we'll parse the prompt to generate a plausible JSON response.
    try:
        if "Intake Agent" in prompt:
            symptoms = prompt.split("Patient Symptoms:")[1].split("Patient History:")[0].strip()
            history = prompt.split("Patient History:")[1].split("Based on this information,")[0].strip()
            plan = ["diagnostics", "treatment", "decision"]
            if "none" in symptoms.lower():
                plan = ["treatment", "decision"]
            return {"symptoms": symptoms, "history": history, "plan": plan}
        elif "Diagnostic Agent" in prompt:
            if "fever" in prompt.lower():
                return {"lab_test": "positive for virus (simulated by QROQ)"}
            elif "pain" in prompt.lower():
                return {"xray_result": "minor fracture (simulated by QROQ)"}
            return {}
        elif "Treatment Agent" in prompt:
            if "positive for virus" in prompt.lower():
                return {"treatment_options": ["Antiviral medication (simulated by QROQ)", "Rest and fluids (simulated by QROQ)"]}
            elif "minor fracture" in prompt.lower():
                return {"treatment_options": ["Splinting (simulated by QROQ)", "Pain medication (simulated by QROQ)"]}
            return {"treatment_options": ["Observation and follow-up (simulated by QROQ)"]}
        elif "Decision Agent" in prompt:
            patient_summary_str = prompt.split("Patient Summary:")[1].split("Diagnostic Results:")[0].strip()
            diagnostics_str = prompt.split("Diagnostic Results:")[1].split("Treatment Options:")[0].strip()
            treatment_options_str = prompt.split("Treatment Options:")[1].split("Provide a JSON object")[0].strip()

            patient_summary = json.loads(patient_summary_str)
            diagnostics = json.loads(diagnostics_str)
            treatment_options = json.loads(treatment_options_str)

            recommendation = "Could not get a specific recommendation from QROQ simulation."
            if "positive for virus" in str(diagnostics).lower():
                recommendation = "Prescribe antiviral medication and recommend rest (QROQ simulated). Schedule a follow-up."
            elif "minor fracture" in str(diagnostics).lower():
                recommendation = "Apply a splint, prescribe pain medication, and refer to orthopedics (QROQ simulated)."
            elif "none" in patient_summary.get("symptoms", "").lower() and "annual check-up" in patient_summary.get("history", "").lower():
                recommendation = "Routine check-up complete. Advise on general wellness (QROQ simulated)."
            elif not treatment_options:
                recommendation = "No immediate treatment required, advise on symptom management and follow up if symptoms worsen (QROQ simulated)."
            else:
                recommendation = "Discuss treatment options with the patient and provide care instructions (QROQ simulated)."

            return {
                "recommendation": recommendation,
                "patient_summary": patient_summary,
                "diagnostics": diagnostics,
                "treatment_options": treatment_options
            }
    except Exception as e:
        print(f"[Simulated QROQ API Error] Could not parse prompt: {e}")
        return {"error": str(e), "message": "Simulated QROQ response failed."}
    time.sleep(0.5) # Simulate API latency
    return {"message": "Simulated QROQ response for an unknown prompt."}

# ================================
# AGENT 1: INTAKE AGENT (QROQ)
# ================================
def intake_agent_qroq(patient_symptoms, patient_history):
    print("\n[Intake Agent (QROQ)] Collecting patient info and planning next steps...")
    prompt = f"""You are an Intake Agent in a hospital. Your task is to collect patient symptoms and history, and then propose a plan for their care. The plan should include 'diagnostics', 'treatment', and 'decision' steps as needed.

Patient Symptoms: {patient_symptoms}
Patient History: {patient_history}

Based on this information, provide a JSON object with the following keys:
'symptoms': string (the provided symptoms)
'history': string (the provided history)
'plan': list of strings (e.g., ["diagnostics", "treatment", "decision"])
"""
    response_data = call_qroq(prompt)
    if "error" in response_data:
        plan = ["diagnostics", "treatment", "decision"] # Fallback if QROQ simulation fails
        print("Using fallback plan due to QROQ simulation error.")
        return {"symptoms": patient_symptoms, "history": patient_history, "plan": plan}

    print("QROQ Intake Agent Response:", response_data)
    time.sleep(1)
    return response_data


# ================================
# AGENT 2: DIAGNOSTIC AGENT (QROQ)
# ================================
def diagnostic_agent_qroq(patient_info):
    print("[Diagnostic Agent (QROQ)] Ordering tests and interpreting results...")
    if "diagnostics" not in patient_info["plan"]:
        return {}

    prompt = f"""You are a Diagnostic Agent. Based on the patient's symptoms and history, and the overall care plan, suggest potential diagnostic tests and their likely results. If no specific tests are needed, indicate that.

Patient Symptoms: {patient_info['symptoms']}
Patient History: {patient_info['history']}
Care Plan: {patient_info['plan']}

Provide a JSON object with potential diagnostic results. For example:
{{ "lab_test": "positive for virus" }} or {{ "xray_result": "minor fracture" }} or {{}}
"""
    response_data = call_qroq(prompt)
    if "error" in response_data:
        print("Using fallback diagnostics due to QROQ simulation error.")
        if "fever" in patient_info["symptoms"].lower():
            return {"lab_test": "positive for virus (simulated fallback)"}
        elif "pain" in patient_info["symptoms"].lower():
            return {"xray_result": "minor fracture (simulated fallback)"}
        else:
            return {"lab_test": "negative (simulated fallback)"}

    print("QROQ Diagnostic Agent Response:", response_data)
    time.sleep(1)
    return response_data


# ================================
# AGENT 3: TREATMENT AGENT (QROQ)
# ================================
def treatment_agent_qroq(patient_info, diagnostics_results):
    print("[Treatment Agent (QROQ)] Suggesting treatment options...")
    if "treatment" not in patient_info["plan"]:
        return []

    prompt = f"""You are a Treatment Agent. Based on the patient's symptoms, history, and diagnostic results, suggest appropriate treatment options.

Patient Symptoms: {patient_info['symptoms']}
Patient History: {patient_info['history']}
Diagnostic Results: {json.dumps(diagnostics_results)}

Provide a JSON object where the key is 'treatment_options' and the value is a list of strings, e.g., {{ "treatment_options": ["Antiviral medication", "Rest and fluids"] }}
"""
    response_data = call_qroq(prompt)
    if "error" in response_data:
        print("Using fallback treatment options due to QROQ simulation error.")
        if "positive for virus" in str(diagnostics_results).lower():
            return ["Antiviral medication (simulated fallback)", "Rest and fluids (simulated fallback)"]
        elif "minor fracture" in str(diagnostics_results).lower():
            return ["Splinting (simulated fallback)", "Pain medication (simulated fallback)"]
        else:
            return ["Observation and follow-up (simulated fallback)"]

    print("QROQ Treatment Agent Response:", response_data)
    time.sleep(1)
    return response_data.get("treatment_options", [])


# ================================
# AGENT 4: DECISION AGENT (QROQ)
# ================================
def decision_agent_qroq(patient_info, diagnostics_results, treatment_options):
    print("[Decision Agent (QROQ)] Reviewing all inputs and providing recommendation...")
    if "decision" not in patient_info["plan"]:
        return {"recommendation": "No final decision step in plan.", "summary": patient_info, "diagnostics": diagnostics_results, "treatment": treatment_options}

    prompt = f"""You are a Decision Agent. Review all available information (patient summary, diagnostics, treatment options) and provide a concise final recommendation to the patient.

Patient Summary: {json.dumps(patient_info)}
Diagnostic Results: {json.dumps(diagnostics_results)}
Treatment Options: {json.dumps(treatment_options)}

Provide a JSON object with the following keys:
'recommendation': string (the final recommendation)
'patient_summary': object (the patient_info passed)
'diagnostics': object (the diagnostics_results passed)
'treatment_options': list of strings (the treatment_options passed)
"""
    response_data = call_qroq(prompt)
    if "error" in response_data:
        print("Using fallback recommendation due to QROQ simulation error.")
        return {
            "patient_summary": patient_info,
            "diagnostics": diagnostics_results,
            "treatment_options": treatment_options,
            "recommendation": "Could not get a recommendation from QROQ. Advise on symptom management and follow up if symptoms worsen (simulated fallback)."
        }

    print("QROQ Decision Agent Response:", response_data)
    time.sleep(1)
    return response_data


# ================================
# MAIN HOSPITAL MULTI-AGENT SYSTEM (QROQ-Powered)
# ================================
def hospital_multi_agent_qroq(patient_query):
    print(f"\nUser Query: {patient_query}")

    symptoms_start = patient_query.find("symptoms:")
    history_start = patient_query.find("history:")

    patient_symptoms = patient_query[symptoms_start + len("symptoms:"):history_start].strip()
    patient_history = patient_query[history_start + len("history:"):].strip()

    # Step 1: Intake Agent (QROQ)
    patient_info = intake_agent_qroq(patient_symptoms, patient_history)

    diagnostics_results = {}
    treatment_options = []
    final_recommendation = {}

    plan = patient_info.get("plan", []) if isinstance(patient_info.get("plan"), list) else ["diagnostics", "treatment", "decision"]

    for step in plan:
        if step == "diagnostics":
            diagnostics_results = diagnostic_agent_qroq(patient_info)
        elif step == "treatment":
            treatment_options_response = treatment_agent_qroq(patient_info, diagnostics_results)
            if isinstance(treatment_options_response, list):
                treatment_options = treatment_options_response
            else:
                treatment_options = treatment_options_response.get("treatment_options", []) if isinstance(treatment_options_response, dict) else []

        elif step == "decision":
            final_recommendation = decision_agent_qroq(patient_info, diagnostics_results, treatment_options)

    return final_recommendation


# RUN SCENARIOS (QROQ-Powered)
print("\n--- Scenario 1 (QROQ): Patient with fever and cough ---")
response1_qroq = hospital_multi_agent_qroq("Patient symptoms: high fever, cough. Patient history: no significant past medical history.")
print("\nFinal Hospital Recommendation (QROQ):")
for key, value in response1_qroq.items():
    print(f"  {key}: {value}")

print("\n--- Scenario 2 (QROQ): Patient with severe wrist pain after a fall ---")
response2_qroq = hospital_multi_agent_qroq("Patient symptoms: severe wrist pain. Patient history: fell off bicycle an hour ago.")
print("\nFinal Hospital Recommendation (QROQ):")
for key, value in response2_qroq.items():
    print(f"  {key}: {value}")

print("\n--- Scenario 3 (QROQ): Routine check-up ---")
response3_qroq = hospital_multi_agent_qroq("Patient symptoms: none. Patient history: annual check-up.")
print("\nFinal Hospital Recommendation (QROQ):")
for key, value in response3_qroq.items():
    print(f"  {key}: {value}")
```

📊 Worker Agents

- Collects competitor data, customer preferences, and demand forecasts.
- Reports: “High demand in Tier‑1 cities, moderate competition.”

- Analyzes budget, ROI, and pricing strategy.
- Reports: “Estimated $10M investment, ROI in 2 years.”

- Assesses staffing needs and training requirements.
- Reports: “Need 50 new hires for customer support and sales.”

```python
import time
import json

# ================================
# WORKER AGENTS
# ================================
def market_research_worker(product, market_region):
    print(f"\n[Market Research Worker] Analyzing market for {product} in {market_region}...")
    time.sleep(1.5)
    report = {
        "demand": "High in Tier-1 cities",
        "competition": "Moderate",
        "customer_preference": "Strong for innovative features"
    }
    print(f"[Market Research Worker] Report: {report}")
    return report

def finance_worker(product, initial_investment):
    print(f"\n[Finance Worker] Analyzing budget and ROI for {product} with initial investment ${initial_investment}M...")
    time.sleep(1.5)
    report = {
        "estimated_investment": f"${initial_investment}M",
        "roi_period": "2-3 years",
        "pricing_strategy": "Premium for early adopters, competitive for mass market"
    }
    print(f"[Finance Worker] Report: {report}")
    return report

def operations_worker(product):
    print(f"\n[Operations Worker] Evaluating supply chain and production for {product}...")
    time.sleep(1.5)
    report = {
        "production_capacity": "Scalable with existing factories",
        "supply_chain_logistics": "Shipping costs are high, local distribution partners needed"
    }
    print(f"[Operations Worker] Report: {report}")
    return report

def legal_worker(product, market_region):
    print(f"\n[Legal Worker] Reviewing compliance and IP for {product} in {market_region}...")
    time.sleep(1.5)
    report = {
        "trademark_status": "Available",
        "regulatory_compliance": "Import laws require specific certifications, data privacy laws need careful adherence"
    }
    print(f"[Legal Worker] Report: {report}")
    return report

def hr_worker():
    print("\n[HR Worker] Assessing staffing needs...")
    time.sleep(1.5)
    report = {
        "new_hires_needed": "50 for customer support, sales, and local operations",
        "training_requirements": "Extensive product knowledge and cultural sensitivity training"
    }
    print(f"[HR Worker] Report: {report}")
    return report

# ================================
# MANAGER AGENT
# ================================
def manager_agent(overall_goal):
    print(f"\n[Manager Agent] Receiving overall goal: '{overall_goal}'")
    print("[Manager Agent] Analyzing goal and assigning tasks...")

    # Dynamic task assignment based on keywords in the goal
    tasks_to_assign = []
    if "feasibility" in overall_goal.lower() or "market" in overall_goal.lower():
        tasks_to_assign.append("market_research")
    if "budget" in overall_goal.lower() or "cost" in overall_goal.lower() or "investment" in overall_goal.lower():
        tasks_to_assign.append("finance")
    if "supply chain" in overall_goal.lower() or "logistics" in overall_goal.lower() or "production" in overall_goal.lower():
        tasks_to_assign.append("operations")
    if "regulations" in overall_goal.lower() or "compliance" in overall_goal.lower() or "legal" in overall_goal.lower():
        tasks_to_assign.append("legal")
    if "staffing" in overall_goal.lower() or "hires" in overall_goal.lower():
        tasks_to_assign.append("hr")

    # Default tasks if no specific keywords are found, or for a general assessment
    if not tasks_to_assign:
        tasks_to_assign = ["market_research", "finance", "operations", "legal", "hr"]

    print(f"[Manager Agent] Assigned tasks: {tasks_to_assign}")
    return tasks_to_assign

# ================================
# EXECUTION SYSTEM
# ================================
def corporate_strategy_multi_agent(overall_goal, product_name="Product Y", market_region="Asia", initial_investment=10):
    print(f"\n--- Initiating Corporate Strategy System for: {overall_goal} ---")

    # Step 1: Manager Agent decides on tasks
    tasks = manager_agent(overall_goal)

    all_reports = {}

    # Step 2: Delegate tasks to Worker Agents
    if "market_research" in tasks:
        all_reports["market_research_report"] = market_research_worker(product_name, market_region)
    if "finance" in tasks:
        all_reports["finance_report"] = finance_worker(product_name, initial_investment)
    if "operations" in tasks:
        all_reports["operations_report"] = operations_worker(product_name)
    if "legal" in tasks:
        all_reports["legal_report"] = legal_worker(product_name, market_region)
    if "hr" in tasks:
        all_reports["hr_report"] = hr_worker()

    # Step 3: Manager Agent compiles and summarizes reports
    print("\n[Manager Agent] Compiling and summarizing all worker reports...")
    final_summary = {
        "overall_goal": overall_goal,
        "product": product_name,
        "market_region": market_region,
        "initial_investment_estimate_M": initial_investment,
        "worker_reports": all_reports,
        "feasibility_conclusion": ""
    }

    # Simple logic for conclusion based on collected reports
    if all_reports.get("market_research_report", {}).get("demand") == "High in Tier-1 cities" and \
       all_reports.get("finance_report", {}).get("roi_period") == "2-3 years" and \
       all_reports.get("legal_report", {}).get("trademark_status") == "Available":
        final_summary["feasibility_conclusion"] = "Feasible with strong potential, but requires careful navigation of logistics and regulations, and significant HR investment."
    else:
        final_summary["feasibility_conclusion"] = "Further investigation needed; potential risks identified. Review specific worker reports for details."

    print("\n--- Final Feasibility Report ---")
    print(json.dumps(final_summary, indent=2))
    return final_summary


# RUN SCENARIOS
print("\n=== Scenario 1: Full Feasibility Evaluation ===")
corporate_strategy_multi_agent("Evaluate feasibility of launching Product Y in Asia.")

print("\n=== Scenario 2: Focus on Legal & Finance ===")
corporate_strategy_multi_agent("Assess legal and financial implications for Product Z launch in Europe.", product_name="Product Z", market_region="Europe", initial_investment=15)

print("\n=== Scenario 3: Operations & HR Challenges ===")
corporate_strategy_multi_agent("Examine operational scalability and staffing needs for new service.", product_name="New Service", initial_investment=5)
```

- Broadcasts: “Data breach detected in customer database. Immediate response required.”

🧑‍⚖️ Agent 7: Decision Agent (Coordinator)

```python
import time
import json

# ================================
# AGENT 1: CRISIS COORDINATOR (BROADCASTER)
# ================================
def crisis_coordinator_agent(crisis_message):
    print(f"\n[Crisis Coordinator] Broadcasting: '{crisis_message}'")
    return crisis_message

# ================================
# AGENT 2: IT SECURITY AGENT
# ================================
def it_security_agent(crisis_message):
    print("[IT Security Agent] Receiving broadcast. Responding...")
    time.sleep(1)
    response = "Isolate affected servers, patch vulnerabilities, start forensic analysis."
    print(f"[IT Security Agent] Response: {response}")
    return {"agent": "IT Security", "response": response}

# ================================
# AGENT 3: COMMUNICATIONS AGENT
# ================================
def communications_agent(crisis_message):
    print("[Communications Agent] Receiving broadcast. Responding...")
    time.sleep(1)
    response = "Draft internal memo, prepare press release, notify stakeholders."
    print(f"[Communications Agent] Response: {response}")
    return {"agent": "Communications", "response": response}

# ================================
# AGENT 4: FINANCE AGENT
# ================================
def finance_agent(crisis_message):
    print("[Finance Agent] Receiving broadcast. Responding...")
    time.sleep(1)
    response = "Estimate financial impact, allocate emergency funds, review insurance coverage."
    print(f"[Finance Agent] Response: {response}")
    return {"agent": "Finance", "response": response}

# ================================
# AGENT 5: LEGAL AGENT
# ================================
def legal_agent(crisis_message):
    print("[Legal Agent] Receiving broadcast. Responding...")
    time.sleep(1)
    response = "Assess regulatory obligations, prepare compliance reports, advise on liability."
    print(f"[Legal Agent] Response: {response}")
    return {"agent": "Legal", "response": response}

# ================================
# AGENT 6: HR AGENT
# ================================
def hr_agent(crisis_message):
    print("[HR Agent] Receiving broadcast. Responding...")
    time.sleep(1)
    response = "Brief employees, provide guidance on handling customer queries, ensure morale support."
    print(f"[HR Agent] Response: {response}")
    return {"agent": "HR", "response": response}

# ================================
# AGENT 7: DECISION AGENT (COORDINATOR)
# ================================
def decision_agent(all_responses):
    print("\n[Decision Agent] Collecting all responses and integrating into a final plan...")
    time.sleep(1.5)

    final_plan_points = []
    for res in all_responses:
        if res["agent"] == "IT Security":
            final_plan_points.append("Servers isolated and forensic analysis initiated.")
        elif res["agent"] == "Communications":
            final_plan_points.append("Internal memo drafted, press release prepared, stakeholders notified.")
        elif res["agent"] == "Finance":
            final_plan_points.append("Financial impact estimated, emergency funds allocated, insurance reviewed.")
        elif res["agent"] == "Legal":
            final_plan_points.append("Regulatory obligations assessed, compliance reports prepared, liability advised.")
        elif res["agent"] == "HR":
            final_plan_points.append("Employees briefed, guidance provided for customer queries, morale support ensured.")

    final_plan = {
        "summary": "Crisis Response Plan",
        "incident": "Data breach detected in customer database.",
        "steps_taken": final_plan_points,
        "full_agent_responses": all_responses
    }
    print("[Decision Agent] Final Crisis Response Plan Compiled.")
    return final_plan

# ================================
# MAIN CRISIS RESPONSE SYSTEM
# ================================
def crisis_response_system(initial_crisis_message):
    print(f"--- Initiating Crisis Response System ---")

    # Agent 1: Crisis Coordinator broadcasts the message
    broadcast_message = crisis_coordinator_agent(initial_crisis_message)

    # Worker Agents receive the broadcast and respond simultaneously (simulated)
    # In a real multi-agent system, this might involve asynchronous calls or message queues
    it_sec_response = it_security_agent(broadcast_message)
    comms_response = communications_agent(broadcast_message)
    finance_response = finance_agent(broadcast_message)
    legal_response = legal_agent(broadcast_message)
    hr_response = hr_agent(broadcast_message)

    all_responses = [
        it_sec_response,
        comms_response,
        finance_response,
        legal_response,
        hr_response
    ]

    # Agent 7: Decision Agent collects and integrates responses
    final_crisis_plan = decision_agent(all_responses)

    print("\n--- Final Crisis Plan ---")
    print(json.dumps(final_crisis_plan, indent=2))
    return final_crisis_plan

# RUN THE SCENARIO
crisis_message = "Data breach detected in customer database. Immediate response required."
full_crisis_plan = crisis_response_system(crisis_message)
```

```python
import requests

# 🔑 PUT YOUR KEY HERE
from google.colab import userdata
API_KEY = userdata.get('groq_api_key')


# ---------------- LLM CALL ---------------- #
def call_llm(prompt):
    url = "https://api.groq.com/openai/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "llama-3.1-8b-instant",
        "messages": [
            {"role": "system", "content": "You are a precise business analyst."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.3,
        "max_tokens": 800
    }

    try:
        res = requests.post(url, headers=headers, json=data, timeout=30)

        if res.status_code != 200:
            print("Status:", res.status_code)
            print("Response:", res.text)

        res.raise_for_status()
        return res.json()["choices"][0]["message"]["content"]

    except Exception as e:
        return f"Error: {e}"


# ---------------- AGENTS ---------------- #

class Agent:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def respond(self, message):
        print(f"\n[{self.name}] thinking...")

        prompt = f"""
You are {self.name}.
Role: {self.role}

Message: {message}

Give 2-3 short actionable steps.
"""
        result = call_llm(prompt)
        print(f"[{self.name}] {result}")
        return result


class DecisionAgent:
    def decide(self, responses):
        print("\n[Decision Agent] combining responses...")

        combined = "\n".join(responses)

        prompt = f"""
Combine these into one final plan:

{combined}
"""
        result = call_llm(prompt)

        print("\n[Final Plan]")
        print(result)


class Coordinator:
    def __init__(self, agents):
        self.agents = agents

    def broadcast(self, message):
        print(f"\n[Coordinator] {message}")
        responses = []

        for agent in self.agents:
            res = agent.respond(message)
            responses.append(res)

        return responses


# ---------------- RUN ---------------- #

if __name__ == "__main__":

    message = "Product X launch in Q3, North America, budget $5M"

    agents = [
        Agent("Marketing Agent", "marketing strategy"),
        Agent("Finance Agent", "budget planning"),
        Agent("Operations Agent", "supply chain"),
        Agent("Legal Agent", "compliance"),
        Agent("HR Agent", "staffing")
    ]

    coordinator = Coordinator(agents)
    decision = DecisionAgent()

    responses = coordinator.broadcast(message)
    decision.decide(responses)
```
