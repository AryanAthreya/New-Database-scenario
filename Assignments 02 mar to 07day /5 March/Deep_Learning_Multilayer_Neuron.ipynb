```python
# Scenario 1: Online Course Completion Prediction
# We want to classify whether a student will complete an online course (1) or drop out (0) based on two features:

# Videos Watched – number of course videos the student watched

# Time Spent on Platform – total minutes spent on the learning platform

# Students who watch more videos and spend more time learning are more likely to complete the course.
# Deep Learning: Multilayer Neural Network Example
# Scenario: Predict Online Course Completion

import numpy as np
# 1. Training Data
X = np.array([
    [2, 15],
    [3, 20],
    [8, 60],
    [9, 75]
])
# Labels: 0 = Dropout, 1 = Complete
y = np.array([[0], [0], [1], [1]])
# Normalize features for stable learning
X = X / np.max(X, axis=0)

# 2. Activation Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def sigmoid_derivative(a):
    return a * (1 - a)

# 3. Network Architecture
input_neurons = 2
hidden_neurons = 3
output_neurons = 1
np.random.seed(42)

weights_input_hidden = np.random.rand(input_neurons, hidden_neurons)
weights_hidden_output = np.random.rand(hidden_neurons, output_neurons)

bias_hidden = np.random.rand(1, hidden_neurons)
bias_output = np.random.rand(1, output_neurons)

# 4. Training (Backpropagation)
learning_rate = 0.1
epochs = 5000
for epoch in range(epochs):
    # Forward Pass
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    predicted_output = sigmoid(final_input)

    # Error
    error = y - predicted_output
    # Backpropagation
    d_output = error * sigmoid_derivative(predicted_output)
    hidden_error = d_output.dot(weights_hidden_output.T)
    d_hidden = hidden_error * sigmoid_derivative(hidden_output)
    # Update Weights
    weights_hidden_output += hidden_output.T.dot(d_output) * learning_rate
    weights_input_hidden += X.T.dot(d_hidden) * learning_rate
    # Update Bias
    bias_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    bias_hidden += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

# 5. Testing the Model
test_student = np.array([[7, 55]])   # Example student
test_student = test_student / np.max(X, axis=0)
hidden_layer = sigmoid(np.dot(test_student, weights_input_hidden) + bias_hidden)
prediction = sigmoid(np.dot(hidden_layer, weights_hidden_output) + bias_output)
print("Completion Probability:", prediction)
if prediction > 0.5:
    print("Student will complete the course")
else:
    print("Student will drop out")
```
