```python
# Employee Attrition Analysis
import pandas as pd
import matplotlib.pyplot as plt

df_attr = pd.read_csv("/content/employee_attrition - Sheet1.csv")

print(df_attr.info())
print("\nMissing Values:\n", df_attr.isnull().sum())
df_attr.head()
```

```python
df_attr["Attrition"].value_counts().plot(kind="bar")

plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()
```

```python
pd.crosstab(df_attr["Department"], df_attr["Attrition"]).plot(kind="bar")

plt.title("Attrition by Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.crosstab(df_attr["Overtime"], df_attr["Attrition"]).plot(kind="bar")

plt.title("Overtime vs Attrition")
plt.show()
```

```python
# 2 Sales Data Analysis
df_sales = pd.read_csv("/content/sample_data1 - Sheet1.csv")

print(df_sales.info())
print("\nMissing Values:\n", df_sales.isnull().sum())
```

```python
df_sales["Monthly_Sales"] = df_sales["Monthly_Sales"].fillna(0)
df_sales["Performance_Rating"] = df_sales["Performance_Rating"].fillna(df_sales["Performance_Rating"].mean())
```

```python
df_sales.groupby("Region")["Monthly_Sales"].sum().plot(kind="bar")

plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales.groupby("Department")["Monthly_Sales"].sum().plot(kind="bar")

plt.title("Sales by Department")
plt.show()
```

```python
# 3 Telecom Churn Analysis
df_churn = pd.read_csv("/content/telecom_churn - Sheet1.csv")

print(df_churn.info())
print("\nMissing Values:\n", df_churn.isnull().sum())
```

```python
df_churn["Churn"].value_counts().plot(kind="bar")

plt.title("Customer Churn Count")
plt.show()
```

```python
pd.crosstab(df_churn["Contract_Type"], df_churn["Churn"]).plot(kind="bar")

plt.title("Churn by Contract Type")
plt.xticks(rotation=45)
plt.show()
```

```python
df_churn.boxplot(column="Monthly_Charges", by="Churn")

plt.suptitle("")
plt.title("Monthly Charges by Churn")
plt.show()
```
