```python
# Visualization
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
salaries = np.random.normal(loc=50000, scale=10000, size=1000)

plt.figure()
plt.hist(salaries, bins=20)
plt.title("Distribution of Employee Salaries")
plt.xlabel("Salary ($)")
plt.ylabel("Number of Employees")
plt.show()
```
