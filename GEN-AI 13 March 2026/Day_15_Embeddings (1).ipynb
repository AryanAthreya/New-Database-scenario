```python
# Simple cosine similarity example
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

a = np.array([0.2,0.8,-0.3,0.5])
b = np.array([0.1,0.9,-0.2,0.4])

print(cosine_similarity(a,b))
```

```python
#example of Dot Product (Fastest Similarity Method)
import numpy as np

a = np.array([1,3,5])
b = np.array([2,4,6])

print(np.dot(a,b))
```

```python

# 🛒 Scenario: E‑Commerce Product Recommendations
# Imagine you’re building a smart recommendation engine for an online store.
# - Problem: Customers often search with phrases like “comfortable running shoes”, but the catalog has items
#  titled “lightweight sneakers for jogging”. Traditional keyword search struggles to connect these.
# - Solution: Use embeddings to capture semantic meaning of product descriptions and customer queries.
import numpy as np

# Step 1: Pretend these are product embeddings
prod1 = np.array([0.7, 0.2, -0.3, 0.6])   # "Lightweight sneakers for jogging"
prod2 = np.array([0.65, 0.25, -0.28, 0.58]) # "Cushioned running trainers"
prod3 = np.array([-0.4, 0.8, 0.1, -0.6])   # "Formal leather office shoes"

# Step 2: Customer search query embedding
query = np.array([0.68, 0.22, -0.29, 0.59])   # "Comfortable running shoes"

# Step 3: Cosine similarity
# Measures semantic similarity between vectors
def cosine_sim(x, y):
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

cos_sim1 = cosine_sim(query, prod1)
cos_sim2 = cosine_sim(query, prod2)
cos_sim3 = cosine_sim(query, prod3)

print("Cosine similarities (higher = more similar):")
print("Product1:", cos_sim1)
print("Product2:", cos_sim2)
print("Product3:", cos_sim3)

# Step 4: Euclidean distance
# Measures geometric distance between vectors
dist1 = np.linalg.norm(query - prod1)
dist2 = np.linalg.norm(query - prod2)
dist3 = np.linalg.norm(query - prod3)

print("\nEuclidean distances (lower = more similar):")
print("Product1:", dist1)
print("Product2:", dist2)
print("Product3:", dist3)

# Step 5: Dot product (normalized)
# Faster alternative to cosine similarity

query_norm = query / np.linalg.norm(query)
prod1_norm = prod1 / np.linalg.norm(prod1)
prod2_norm = prod2 / np.linalg.norm(prod2)
prod3_norm = prod3 / np.linalg.norm(prod3)

dot1 = np.dot(query_norm, prod1_norm)
dot2 = np.dot(query_norm, prod2_norm)
dot3 = np.dot(query_norm, prod3_norm)

print("\nDot products (normalized):")
print("Product1:", dot1)
print("Product2:", dot2)
print("Product3:", dot3)
```

```python
# Scenario: Hospital FAQ Assistant
# A hospital wants to build a chatbot that helps patients quickly find answers to common questions.
# - Problem: Patients ask questions in different ways. For example:
# - “How do I book an appointment?”
# - “What’s the process to schedule a doctor visit?”
# - “Can I see a physician tomorrow?”
# Traditional keyword search might miss these connections.
# - Solution: Convert FAQs and patient queries into embeddings, then measure similarity to find the closest match.

# 👩‍💻 Code in Context
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# Step 1: Pretend these are FAQ embeddings
# (In reality, you'd generate them with a model like MiniLM or mpnet)
# Each vector is a simplified 4-dimensional embedding
faq1 = np.array([0.2, 0.8, -0.3, 0.5])   # "How do I book an appointment?"
faq2 = np.array([0.1, 0.9, -0.2, 0.4])   # "Process to schedule a doctor visit"
faq3 = np.array([-0.7, 0.1, 0.9, -0.5])  # "Visiting hours for patients"

# -----------------------------
# Step 2: Student query embedding
# Again, simplified for demonstration
query = np.array([0.19, 0.82, -0.25, 0.48])  # "Can I see a physician tomorrow?"

# -----------------------------
# Step 3: Cosine similarity
# Measures the angle between vectors (direction matters, not magnitude)
def cosine_sim(x, y):
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

cos_sim1 = cosine_sim(query, faq1)
cos_sim2 = cosine_sim(query, faq2)
cos_sim3 = cosine_sim(query, faq3)

print("Cosine similarities (higher = more similar):")
print("FAQ1:", cos_sim1)
print("FAQ2:", cos_sim2)
print("FAQ3:", cos_sim3)

# -----------------------------
# Step 4: Euclidean distance
# Measures actual geometric distance (lower = more similar)
dist1 = np.linalg.norm(query - faq1)
dist2 = np.linalg.norm(query - faq2)
dist3 = np.linalg.norm(query - faq3)

print("\nEuclidean distances (lower = more similar):")
print("FAQ1:", dist1)
print("FAQ2:", dist2)
print("FAQ3:", dist3)

# -----------------------------
# Step 5: Dot product (normalized)
# Fastest metric, equivalent to cosine if vectors are normalized
query_norm = query / np.linalg.norm(query)
faq1_norm = faq1 / np.linalg.norm(faq1)
faq2_norm = faq2 / np.linalg.norm(faq2)
faq3_norm = faq3 / np.linalg.norm(faq3)

dot1 = np.dot(query_norm, faq1_norm)
dot2 = np.dot(query_norm, faq2_norm)
dot3 = np.dot(query_norm, faq3_norm)

print("\nDot products (normalized, same as cosine):")
print("FAQ1:", dot1)
print("FAQ2:", dot2)
print("FAQ3:", dot3)

# 📊 What Happens
# - Cosine similarity will show that the query is closest to FAQ1 and FAQ2 (both about appointments).
# - Euclidean distance will also confirm closeness, but it’s more sensitive to vector magnitude.
# - Dot product (if normalized) would give the same ranking as cosine, but faster.

# 🎯 Teaching Point
# This scenario shows how similarity metrics decide which FAQ answer to return:
# - Cosine similarity → best for text queries (robust to phrasing differences).
# - Euclidean distance → useful if absolute position matters (e.g., clustering patient queries).
# - Dot product → fastest option when vectors are normalized, ideal for large‑scale systems.

# 👉 In short: the code isn’t just math — it’s the engine behind semantic search and retrieval systems that make chatbots, recommendation engines, and assistants truly useful.
# Would you like me to design a business scenario (like customer support or HR policy search) next, so you can use it directly in your corporate training sessions?
```

```python
# Scenario: Hospital Patient Record Search
# - Problem: Doctors and nurses often type queries like “heart attack” or “chest pain emergency”, but the hospital’s electronic health record (EHR)
# system stores diagnoses as “acute myocardial infarction”. Traditional keyword search fails because medical staff use everyday language while records
# use clinical terminology.
# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that “heart attack” and “acute myocardial infarction” are related concepts.
# This allows the search system to return the right patient records even when the query wording doesn’t match the stored terminology.
# import numpy as np

# Install required libraries (run once)
# pip install sentence-transformers scikit-learn

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# -----------------------------
# Load embedding model
# -----------------------------
model = SentenceTransformer('all-MiniLM-L6-v2')

# -----------------------------
# Hospital patient records
# -----------------------------
patient_records = [
    "Patient diagnosed with acute myocardial infarction and admitted to ICU",
    "Patient suffering from hypertension and high blood pressure",
    "Diagnosis shows bronchial asthma and breathing difficulty",
    "Patient treated for diabetes mellitus type 2",
    "Case of cerebrovascular accident (stroke) with neurological symptoms"
]

# -----------------------------
# Convert records into embeddings
# -----------------------------
record_embeddings = model.encode(patient_records)

# -----------------------------
# Doctor / Nurse search query
# -----------------------------
query = "heart attack emergency"

query_embedding = model.encode([query])

# -----------------------------
# Calculate similarity
# -----------------------------
similarities = cosine_similarity(query_embedding, record_embeddings)[0]

# -----------------------------
# Get best matching record
# -----------------------------
best_match_index = np.argmax(similarities)

print("Search Query:", query)
print("Most Relevant Patient Record:")
print(patient_records[best_match_index])
print("Similarity Score:", similarities[best_match_index])
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Patient records stored in hospital database
records = [
    "Acute myocardial infarction",
    "Hypertension",
    "Cardiac arrest",
    "Diabetes mellitus"
]

# Convert records into embeddings (vectors)
record_vectors = model.encode(records)

# Doctor query
query = "heart attack"

# Convert query into embedding
query_vector = model.encode([query])

# Calculate similarity between query and records
similarities = model.similarity(query_vector, record_vectors)

# Print results
for record, score in zip(records, similarities[0]):
    print(f"{record} -> Similarity Score: {score:.3f}")

# Find best match
best_index = similarities.argmax()
print("\nMost Relevant Record:", records[best_index])
```

```python
# Scenario: E‑Commerce Product Search
# - Problem: Shoppers often type queries like “cheap running shoes” or “affordable sneakers for jogging”, but the catalog lists products as “Budget-friendly athletic footwear”. Traditional keyword search misses these connections because customers use everyday language while product listings use marketing terms.
# - Solution: Use embeddings to capture semantic meaning. Embeddings understand that “cheap running shoes” and “budget-friendly athletic footwear” are related concepts. This ensures customers find the right products even when their wording doesn’t match the catalog descriptions.
from sentence_transformers import SentenceTransformer, util

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# -----------------------------
# Step 1: Product catalog
products = [
    "Budget-friendly athletic footwear",
    "Premium performance running sneakers",
    "Luxury leather office shoes"
]

# Convert product descriptions into embeddings
product_embeddings = model.encode(products, convert_to_tensor=True)

# -----------------------------
# Step 2: Customer query
query = "cheap running shoes"

query_embedding = model.encode(query, convert_to_tensor=True)

# -----------------------------
# Step 3: Semantic similarity search
similarities = util.cos_sim(query_embedding, product_embeddings)

# Find best match
best_match_index = similarities.argmax()

print("Customer Query:", query)
print("Best Matching Product:", products[best_match_index])
print("Similarity Scores:", similarities)
```
