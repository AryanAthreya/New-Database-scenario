```python
import matplotlib.pyplot as plt

# Data
months = ["January", "February", "March", "April"]
attendance = [85, 90, 95, 88]

# Create line chart
plt.figure()
plt.plot(months, attendance, marker='o')

# Add title and labels
plt.title("Monthly Student Attendance Trend")
plt.xlabel("Month")
plt.ylabel("Average Number of Students Present")

plt.grid(True)
plt.show()
```
