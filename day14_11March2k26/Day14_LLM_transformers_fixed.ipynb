### Practice 3: Research Assistant – BERT Tokenization

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = "Deep learning models are powerful"
tokens = tokenizer.tokenize(text)
print(tokens)
```

### Practice 4: Healthcare Chatbot Tokenization

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
text = "I have chest pain and shortness of breath"
tokens = tokenizer.tokenize(text)
print("Original:", text)
print("Tokens:", tokens)
```

### Practice 5: Temperature Control (Creativity)

```python
from transformers import pipeline
generator = pipeline("text-generation", model="gpt2")
prompt = "The future of Artificial Intelligence"

print("Low temperature (more predictable):\n")
print(generator(prompt, max_length=50, temperature=0.2)[0]['generated_text'])

print("\nHigh temperature (more creative):\n")
print(generator(prompt, max_length=50, temperature=0.8)[0]['generated_text'])
```

### Practice 6: Scriptwriting Assistant (Temperature Demo)

```python
from transformers import pipeline
generator = pipeline("text-generation", model="gpt2")
prompt = "The future of Artificial Intelligence"

print("Temperature = 0.2 (formal):")
print(generator(prompt, max_length=60, temperature=0.2)[0]['generated_text'])

print("\nTemperature = 0.8 (creative):")
print(generator(prompt, max_length=60, temperature=0.8)[0]['generated_text'])
```

### Practice 9–10: Model Comparison (GPT-2 vs DistilGPT-2)

```python
from transformers import pipeline

gpt2 = pipeline("text-generation", model="gpt2")
distil = pipeline("text-generation", model="distilgpt2")

prompt = "Artificial Intelligence will transform"

print("GPT-2:")
print(gpt2(prompt, max_length=50)[0]['generated_text'])
print("\nDistilGPT-2 (faster):")
print(distil(prompt, max_length=50)[0]['generated_text'])
```
