```python
# Data_Transformation
import pandas as pd

df = pd.read_csv("/content/Dataset  - Sheet1 (1).csv")

print(df)
```

```python
exceptional = df[df["Score"] > 90]
print("Exceptional Performers:\n")
print(exceptional)
```

```python
sales_dept = df[df["Department"] == "Sales"]
print("Sales Department Records:\n")
print(sales_dept)
```

```python
ranked = df.sort_values("Score", ascending=False)
print("Performance Ranking:\n")
print(ranked)
```

```python
dept_rank = df.sort_values(["Department", "Score"], ascending=[True, False])
print("Department-wise Ranking:\n")
print(dept_rank)
```

```python
avg_dept_score = df.groupby("Department")["Score"].mean()
print("Average Score by Department:\n")
print(avg_dept_score)
```

```python
max_quarter_score = df.groupby("Quarter")["Score"].max()
print("Maximum Score by Quarter:\n")
print(max_quarter_score)
```
